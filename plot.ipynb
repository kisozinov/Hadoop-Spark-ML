```python
import json
```

```python
contents = open("metrics.json", "r").read() 
data = [json.loads(str(item)) for item in contents.strip().split('\n')]
results = {list(item.keys())[0]: list(item.values())[0] for item in data}
```

```python
results
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
for key, value in results.items():
    plt.plot(value['time'], label=key)
plt.title('Time for Each Experiment')
plt.xlabel('Stages')
plt.ylabel('Cumulative Time (seconds)')
plt.legend()

plt.subplot(1, 2, 2)
for key, value in results.items():
    plt.plot(value['memory'], label=key)
plt.title('Memory Usage for Each Experiment')
plt.xlabel('Stages')
plt.ylabel('Memory Usage (MB)')
plt.legend()

plt.tight_layout()
plt.show()
```
